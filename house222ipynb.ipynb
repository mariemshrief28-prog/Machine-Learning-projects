```python
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('HousingData.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

```python
x=df.drop(['MEDV','CHAS'],axis=1)
x.head()
```

```python
y=df['MEDV']
y
```

```python
for colms in x.columns:
  plt.hist(x[colms])
  plt.xlabel(xlabel=colms)
  plt.show()
```

```python
for colms in x.columns:
  plt.scatter(x[colms],y)
  plt.xlabel(xlabel=colms)
  plt.ylabel(ylabel="MEDV")
  plt.show()
```

```python
import matplotlib.pyplot as plt
x.boxplot()
plt.show()
```

```python
for colms in x.columns:
  q1=x[colms].quantile(0.25)
  q3=x[colms].quantile(0.75)

  iqr=q3-q1

  upper=q3+3*iqr
  lower=q1-3*iqr

  x[colms]=x[colms].loc[(x[colms]<upper)&(x[colms]>lower)]
```

```python
import matplotlib.pyplot as plt
x.boxplot()
plt.show()
```

```python
x=x.drop(['ZN','B','CRIM','RAD'],axis=1)
```

```python
for colm in x.columns:
  print(colm,(x[colm]==0).sum())
```

```python
for colm in x.columns:
  print(colm,(x[colm]==0).sum())
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler
sd= StandardScaler()
lis=x.columns
x[lis]=sd.fit_transform(x[lis])
```

```python
x.head()
```

```python
for colm in x.columns:
  print(x[colm].isnull().sum())
```

```python
for colm in x.columns:
  x[colm]=x[colm].fillna(x[colm].mean())
  print(x[colm].isnull().sum())
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,
                                               y,
                                               train_size=.75,
                                               random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
lr=LinearRegression()
lr.fit(x_train,y_train)
y_predict = lr.predict(x_test)
r2 = metrics.r2_score(y_test,y_predict)
mae = metrics.mean_absolute_error(y_test, y_predict)
mse = metrics.mean_squared_error(y_test, y_predict)
print(f" R2 Score = {r2*100:.2f} %\n MAE = {mae:.2f}\n MSE = {mse:.2f} ")
```

```python
from sklearn.svm import SVR
from sklearn import metrics
svr=SVR(kernel='rbf')
svr.fit(x_train,y_train)
y_predict=svr.predict(x_test)
r2 = metrics.r2_score(y_test,y_predict)
mae = metrics.mean_absolute_error(y_test, y_predict)
mse = metrics.mean_squared_error(y_test, y_predict)
print(f" R2 Score = {r2*100:.2f} %\n MAE = {mae:.2f}\n MSE = {mse:.2f} ")
```
