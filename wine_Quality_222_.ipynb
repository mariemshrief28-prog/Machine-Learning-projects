```python
x=df.drop(['quality','Id','residual sugar','pH','free sulfur dioxide'],axis=1)
x.head()
```

```python
plt.hist(y)
plt.show()
```

```python
from imblearn.over_sampling import RandomOverSampler

oversample = RandomOverSampler(sampling_strategy='all')
x_over, y_over = oversample.fit_resample(x, y)

print('x_over:\n', x_over)
print('y_over:\n', y_over)
```

```python
plt.hist(y_over)
plt.show()
```

```python
for colms in x_over.columns:
  q1=x_over[colms].quantile(0.25)
  q3=x_over[colms].quantile(0.75)

  iqr=q3-q1

  upper=q3+0.5*iqr
  lower=q1-0.5*iqr

  x_over[colms]=x_over[colms].loc[(x_over[colms]<upper)&(x_over[colms]>lower)]
```

```python
import matplotlib.pyplot as plt
x_over.boxplot()
plt.show()
```

```python
for colms in x_over.columns:
  x_over[colms]= (x_over[colms]-x_over[colms].mean())/x_over[colms].std()
```

```python
from sklearn.svm import SVC
from sklearn import metrics

svm=SVC(kernel='rbf')
svm.fit(x_train,y_train)
y_predict=svm.predict(x_test)
acc=metrics.accuracy_score(y_test,y_predict)
PRE= metrics.precision_score(y_test, y_predict, average='micro')
REC= metrics.recall_score(y_test, y_predict, average='micro')
print(f" accuacry= {acc*100:.2f} \n Pre= {PRE*100:.2f}\n Rec= {REC*100:.2f} ")
```
