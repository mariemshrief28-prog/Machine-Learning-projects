```python
q1=x['Pregnancies'].quantile(0.25)
q3=x['Pregnancies'].quantile(0.75)
iqr=q3-q1
upper=q3+ 1*iqr
lower=q1- 1*iqr
x['Pregnancies']=x['Pregnancies'].loc[(x['Pregnancies']<upper)&(x['Pregnancies']>lower)]

q1=x['Glucose'].quantile(0.25)
q3=x['Glucose'].quantile(0.75)
iqr=q3-q1
upper=q3+ 1.5*iqr
lower=q1- 1.5*iqr
x['Glucose']=x['Glucose'].loc[(x['Glucose']<upper)&(x['Glucose']>lower)]

q1=x['BloodPressure'].quantile(0.25)
q3=x['BloodPressure'].quantile(0.75)
iqr=q3-q1
upper=q3+ 1*iqr
lower=q1- 1*iqr
x['BloodPressure']=x['BloodPressure'].loc[(x['BloodPressure']<upper)&(x['BloodPressure']>lower)]

q1=x['Insulin'].quantile(0.25)
q3=x['Insulin'].quantile(0.75)
iqr=q3-q1
upper=q3+ 1*iqr
lower=q1- 1*iqr
x['Insulin']=x['Insulin'].loc[(x['Insulin']<upper)&(x['Insulin']>lower)]

q1=x['BMI'].quantile(0.25)
q3=x['BMI'].quantile(0.75)
iqr=q3-q1
upper=q3+ 1.5*iqr
lower=q1- 1.5*iqr
x['BMI']=x['BMI'].loc[(x['BMI']<upper)&(x['BMI']>lower)]


q1=x['DiabetesPedigreeFunction'].quantile(0.25)
q3=x['DiabetesPedigreeFunction'].quantile(0.75)
iqr=q3-q1
upper=q3+ 1*iqr
lower=q1- 1*iqr
x['DiabetesPedigreeFunction']=x['DiabetesPedigreeFunction'].loc[(x['DiabetesPedigreeFunction']<upper)&(x['DiabetesPedigreeFunction']>lower)]



q1=x['Age'].quantile(0.25)
q3=x['Age'].quantile(0.75)
iqr=q3-q1
upper=q3+ 1*iqr
lower=q1- 1*iqr
x['Age']=x['Age'].loc[(x['Age']<upper)&(x['Age']>lower)]

```

```python
from imblearn.over_sampling import RandomOverSampler
rs=RandomOverSampler(sampling_strategy='minority')
X,Y=rs.fit_resample(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,train_size=.8,random_state=8)
```

```python
from collections import Counter

print(Counter(y_train))
```

```python
from sklearn import svm

model=svm.SVC()
model.fit(x_train,y_train)

y_predict= model.predict(x_test)

for x in range(len(y_predict)):
  print(y_predict[x],y_test.iloc[x])
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score
acc=accuracy_score(y_test,y_predict)
prec=precision_score(y_test,y_predict,average = 'weighted')
rec=recall_score(y_test,y_predict,average = 'weighted')
print(acc,prec,rec)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(max_iter=10000, random_state=0)
clf.fit(x_train, y_train)

acc = accuracy_score(y_test, clf.predict(x_test)) * 100
print(f"Logistic Regression model accuracy: {acc:.2f}%")
```
