```python

class seq():
  def __init__(self):
    super().__init__()
    self.l1=nn.Linear(in_features=10,out_features=8)
    self.l2=nn.Linear(in_features=8,out_features=8)
    self.l3=nn.Linear(in_features=8,out_features=2)
    self.l5=nn.Sigmoid()
    self.l4=nn.SiLU()
  def forward(self,X: torch.Tensor):
    return self.l3(self.l4(self.l2(self.l5(self.l1(X)))))
```

```python
epochs=2000
loss_train=[]
nepoch=[]
loss_test=[]
loss_fun=nn.CrossEntropyLoss()
optimizer=torch.optim.SGD(params=model_0.parameters(),lr=.3)
```

```python
for epoch in range(epochs):
  model_0.train()
  train_logit=model_0(torch.from_numpy(x_train_tensor.numpy()).float())
  y_train_torch = torch.from_numpy(y_train_tensor.numpy()).long()

  train_pred=torch.softmax(train_logit,dim=1).argmax(dim=1)
  loss=loss_fun(train_logit,y_train_torch)
  loss_train.append(loss)
  nepoch.append(epoch)
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  model_0.eval()
  with torch.inference_mode():
    test_logit=model_0(torch.from_numpy(x_test_tensor.numpy()).float())
    y_test_torch = torch.from_numpy(y_test_tensor.  ()).long()

    test_pred=torch.softmax(test_logit,dim=1).argmax(dim=1)
    test_loss=loss_fun(test_logit,y_test_torch)
    loss_test.append(test_loss)

    if epoch%100==0:
      print(f"Epoch: {epoch}, Train Loss: {loss.item():.4f}, Test Loss: {test_loss.item():.4f}")
```

```python
plt.figure(figsize=(10, 7))
plt.plot(nepoch, [l.item() for l in loss_train], label='Training Loss')
plt.plot(nepoch, [l.item() for l in loss_test], label='Test Loss')
plt.title('Training and Test Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
def plot_decision_boundary(model, X, y, title, feature_x_idx=0, feature_y_idx=1, base_features_mean=None, resolution=200):

    if isinstance(X, torch.Tensor):
        X_np = X.detach().cpu().numpy()
    else:
        X_np = X
    if isinstance(y, torch.Tensor):
        y_np = y.detach().cpu().numpy()
    else:
        y_np = y

    x_min, x_max = X_np[:, feature_x_idx].min() - 0.1, X_np[:, feature_x_idx].max() + 0.1
    y_min, y_max = X_np[:, feature_y_idx].min() - 0.1, X_np[:, feature_y_idx].max() + 0.1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, resolution),
                         np.linspace(y_min, y_max, resolution))

    if base_features_mean is None:
        base_features_mean = X_np.mean(axis=0)

    full_grid_data = np.tile(base_features_mean, (xx.size, 1))

    full_grid_data[:, feature_x_idx] = xx.ravel()
    full_grid_data[:, feature_y_idx] = yy.ravel()

    grid_torch = torch.tensor(full_grid_data, dtype=torch.float32)

    with torch.no_grad():
        preds = model(grid_torch)
        if preds.shape[1] > 1:
            preds = torch.argmax(preds, dim=1)
        else:
            preds = torch.round(torch.sigmoid(preds))
    Z = preds.reshape(xx.shape).cpu().numpy()

    plt.figure(figsize=(7, 6))
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.RdYlBu)
    plt.scatter(X_np[:, feature_x_idx], X_np[:, feature_y_idx], c=y_np, cmap=plt.cm.RdYlBu, edgecolor="k", s=40)
    plt.title(title)
    plt.xlabel(f"Feature {feature_x_idx}")
    plt.ylabel(f"Feature {feature_y_idx}")
    plt.show()
```

```python
mean_x_train_features = x_train_np.mean(axis=0)

plot_decision_boundary(model_0,
                       torch.from_numpy(x_train_tensor.numpy()).float(),
                       torch.from_numpy(y_train_tensor.numpy()).long(),
                       "train",
                       feature_x_idx=0, feature_y_idx=1,
                       base_features_mean=mean_x_train_features)

plot_decision_boundary(model_0,
                       torch.from_numpy(x_test_tensor.numpy()).float(),
                       torch.from_numpy(y_test_tensor.numpy()).long(),
                       "test",
                       feature_x_idx=0, feature_y_idx=1,
                       base_features_mean=mean_x_train_features)
```
