```python
noise=x[['Rented Bike Count','Wind speed (m/s)']]
print(noise)
```

```python
q1=x['Rented Bike Count'].quantile(0.25)
q3=x['Rented Bike Count'].quantile(0.75)
iqr=q3-q1
upper=q3+ 1*iqr
lower= q1- 1*iqr
x['Rented Bike Count']=x['Rented Bike Count'].loc[(x['Rented Bike Count']<upper)&(x['Rented Bike Count']>lower)]
```

```python
q1=x['Wind speed (m/s)'].quantile(0.25)
q3=x['Wind speed (m/s)'].quantile(0.75)
iqr=q3-q1
upper=q3+ 1.5*iqr
lower=q1- 1.5*iqr
x['Wind speed (m/s)']=x['Wind speed (m/s)'].loc[(x['Wind speed (m/s)']<upper)&(x['Wind speed (m/s)']>lower)]
```

```python
null1=x['Rented Bike Count'].mean

x['Rented Bike Count'].fillna(null1)

null2=x['Wind speed (m/s)'].mean
x['Wind speed (m/s)'].fillna(null2)
```

```python
from imblearn.over_sampling import RandomOverSampler
rs=RandomOverSampler(sampling_strategy='minority')
X,Y=rs.fit_resample(x,y)
```

```python
from collections import Counter

print(Counter(Y))
```

```python
x_train,x_test,y_train,y_test= train_test_split(X,Y,train_size=.75,random_state=10)
```

```python
from collections import Counter

print(Counter(y_train))
```

```python
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier, plot_tree
clf_gini = DecisionTreeClassifier(criterion="gini", random_state=80, max_depth=7, min_samples_leaf=8)
clf_gini.fit(x_train, y_train)
y_pred = clf_gini.predict(x_test)
print("Predicted Values:\n", y_pred)

print("\nAccuracy:", metrics.accuracy_score(y_test, y_pred) * 100)
print("\nClassification Report:\n", metrics.precision_score(y_test, y_pred, average='micro'))
print("\nClassification Report:\n", metrics.classification_report(y_test, y_pred))
```
