```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Titanic-Dataset.csv')
df.head()
```

```python
import sklearn
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

df['Name']=le.fit_transform(df['Name']).astype(int)
df['Cabin']=le.fit_transform(df['Cabin']).astype(int)
df['Sex']=le.fit_transform(df['Sex']).astype(int)
df['Ticket']=le.fit_transform(df['Ticket']).astype(int)
df['Embarked']=le.fit_transform(df['Embarked']).astype(int)

df.head()
```

```python
for colms in x.columns:
  (x[colms]==0).sum()
  print({colms},(x[colms]==0).sum())
  x[colms] = x[colms].replace(0, x[colms].mean())
  print({colms},(x[colms]==0).sum())
```

```python
x.boxplot()
```

```python
for colms in x.columns:
  plt.hist(x[colms].dropna())
  plt.xlabel(xlabel=colms)
  plt.figure(figsize=(17,22))
  plt.show()
```

```python
for colms in x.columns:
  q1=x[colms].quantile(0.25)
  q3=x[colms].quantile(0.75)

  iqr=q3-q1

  upper=q3+1.5*iqr
  lower=q1-1.5*iqr

  x[colms]=x[colms].loc[(x[colms]<upper)&(x[colms]>lower)]
```
